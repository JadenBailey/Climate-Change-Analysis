## Section 1: Sampling CO2 Levels

For this section, I will be sampling the carbon dioxide levels from 1959 and 2016. The reason I chose these two years is because these two years are the earliest and latest years that have full data all throughout the year. 1958 and 2017 contain missing data that I may find later.

```python
sampleOne = data[data.Year == 1959]
sampleTwo = data[data.Year == 2016]
```

```python
sampleOne
```

```python
sampleTwo
```

```python
plt.plot(sampleOne.Month, sampleOne['Carbon Dioxide (ppm)'])
plt.plot(sampleTwo.Month, sampleTwo['Carbon Dioxide (ppm)'])
plt.legend(['1959', '2016'])
plt.xlabel('Month')
plt.ylabel('Carbon Dioxide (ppm)')
plt.show()
```

Based off of the chart above, carbon dioxide levels have dramatically increased over the 57 year timespan from below 320 ppm (parts per million) to over 400 ppm and rising.
